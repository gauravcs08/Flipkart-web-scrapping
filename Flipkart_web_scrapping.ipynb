```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.flipkart.com/television-store?otracker=nmenu_sub_TVs%20and%20Appliances_0_Televisions&otracker=nmenu_sub_TVs%20%26%20Appliances_0_Television'
```

```python
response=requests.get(url)
data=response.text
soup=BeautifulSoup(data,'html.parser')
views=soup.find_all('a',{'class':'_2AkmmA _1eFTEo'})
tv_list={}
tv_no=0
```

```python
for view in views:
    url=view.get('href')
    url='https://www.flipkart.com'+url
    response=requests.get(url)
    data1=response.text
    soup=BeautifulSoup(data1,'html.parser')
    jobs=soup.find_all('div',{'class':'_1-2Iqu row'})
    for job in jobs:
        name1=job.find('div',{'class':'_3wU53n'})
        name=name1.text if name1 else 'N/A'
        price1=job.find('div',{'class':'_1vC4OE _2rQ-NK'})
        price=price1.text if price1 else 'N/A'
        Rating1=job.find('div',{'class':'hGSR34'})
        Rating=Rating1.text if Rating1 else 'N/A'
        discount1=job.find('div',{'class':'VGWI6T'})
        discount=discount1.text if discount1 else 'N/A'
        tv_no+=1
        tv_list[tv_no]=[name,price,Rating,discount]
    
```

```python
tv_list
```

```python
k=pd.DataFrame.from_dict(tv_list,orient='index',columns=['Product_Name','Current_Price','Ratings','Discounts'])
```
